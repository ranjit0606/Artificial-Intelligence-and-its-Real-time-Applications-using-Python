```python
import spacy
nlp= spacy.load('en_core_web_sm')
```

```python
doc1=nlp(u'i am runner running ina race because i love to run since i ran today')
for token in doc1:
  print(token.text, '\t', token.pos_,'\t',token.lemma,'\t',token.lemma_)
```

```python
def show_lemmas(text):
  for token in text:
    print(f'{token.text:{12}} {token.pos_:{6}} {token.lemma:<{22}} {token.lemma_}')
```

```python
doc2=nlp(u'i saw eighteen mice today!')
show_lemmas(doc2)
```

```python
doc2=nlp(u'saravana is a play boy!')
show_lemmas(doc2)
```

```python
doc2=nlp(u'sam is first rank holder in college')
show_lemmas(doc2)
```
