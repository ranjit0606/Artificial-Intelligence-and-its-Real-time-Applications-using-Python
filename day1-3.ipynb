```python
text = "my telephone number is 409-555-1234"
```

```python
phone= re.search(r'\d\d\d-\d\d\d-\d\d\d\d',text)
```

```python
phone.group()
```

```python
re.search(r'\d{3}-\d{3}-\{4}',text)
```

```python
phone_pattern=re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
results= re.search(phone_pattern,text)
```

```python
results.group()
```
