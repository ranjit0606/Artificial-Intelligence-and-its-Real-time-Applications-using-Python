```python
#min width alignment padding
library=[('author','topic','pages'),('twain','rafting',100),('feyman','physics',200),('hamilton','mythology',144)]
for book in library:
    print(f'{book[0]:{10}} {book[1]:{8}} {book[2]:{7}}')
```
